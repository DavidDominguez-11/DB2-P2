```python
from faker import Faker
import csv
import random
import os

fake = Faker()

NUM_USERS = 1500
NUM_ARTISTS = 500
NUM_GENRES = 30
NUM_SONGS = 1500
NUM_PLAYLISTS = 1000
NUM_POSTS = 1000

genres_list = ["rock", "pop", "jazz", "hiphop", "electronic", "latin"]

os.makedirs("data", exist_ok=True)

# USERS 
print("Generando usuarios")
with open("data/users.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user_id","username","email","fecha_registro","premium","generos"])

    for i in range(1, NUM_USERS+1):
        writer.writerow([
            i,
            fake.user_name(),
            fake.email(),
            fake.date_this_decade(),
            random.choice([True, False]),
            ",".join(random.sample(genres_list, k=random.randint(1,3)))
        ])
print(f"{NUM_USERS} usuarios generados en data/users.csv")

# ARTISTS 
print("Generando artistas")
with open("data/artists.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["artist_id","nombre","pais","genero_principal","activo","anios_activo"])

    for i in range(1, NUM_ARTISTS+1):
        writer.writerow([
            i,
            fake.name(),
            fake.country(),
            random.choice(genres_list),
            random.choice([True, False]),
            random.randint(1, 30)
        ])
print(f"{NUM_ARTISTS} artistas generados en data/artists.csv")

# GENRES 
print("Generando generos")
with open("data/genres.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["genre_id","nombre","descripcion","popularidad","origen","activo"])

    for i in range(1, NUM_GENRES+1):
        writer.writerow([
            i,
            random.choice(genres_list),
            fake.sentence(),
            random.randint(1,100),
            fake.country(),
            True
        ])
print(f"{NUM_GENRES} generos generados en data/genres.csv")

# SONGS 
print("Generando canciones")
with open("data/songs.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["song_id","titulo","duracion","fecha_lanzamiento","popularidad","idiomas"])

    for i in range(1, NUM_SONGS+1):
        writer.writerow([
            i,
            fake.word(),
            round(random.uniform(2,5),2),
            fake.date_this_decade(),
            random.randint(1,100),
            "english,spanish"
        ])
print(f"{NUM_SONGS} canciones generadas en data/songs.csv")

# PLAYLISTS 
print("Generando playlists")
with open("data/playlists.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["playlist_id","nombre","descripcion","fecha_creacion","publica","numero_canciones"])

    for i in range(1, NUM_PLAYLISTS+1):
        writer.writerow([
            i,
            fake.word(),
            fake.sentence(),
            fake.date_this_decade(),
            random.choice([True, False]),
            random.randint(5,50)
        ])
print(f"{NUM_PLAYLISTS} playlists generadas en data/playlists.csv")

# POSTS 
print("Generando posts")
with open("data/posts.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["post_id","caption","fecha","tipo","privacidad","hashtags"])

    for i in range(1, NUM_POSTS+1):
        writer.writerow([
            i,
            fake.sentence(),
            fake.date_this_decade(),
            random.choice(["music","share","recommend"]),
            random.choice(["public","private"]),
            "#music,#fun"
        ])
print(f"{NUM_POSTS} posts generados en data/posts.csv")

print("Todos los CSV de nodos generados correctamente en la carpeta data/")
```

```python
# FOLLOWS (User → User)
print("Generando follows")

with open("data/follows.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user1","user2","fecha","notificaciones","cercania"])

    for _ in range(3000):
        u1 = random.randint(1, NUM_USERS)
        u2 = random.randint(1, NUM_USERS)
        if u1 != u2:
            writer.writerow([
                u1,
                u2,
                fake.date_this_decade(),
                random.choice([True, False]),
                round(random.random(),2)
            ])
print("3000 follows generados en data/follows.csv")
```

```python
# LISTENED (User → Song)
print("Generando listened")
with open("data/listened.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user_id","song_id","fecha","duracion","completado"])

    for _ in range(5000):
        writer.writerow([
            random.randint(1, NUM_USERS),
            random.randint(1, NUM_SONGS),
            fake.date_this_decade(),
            round(random.uniform(1,5),2),
            random.choice([True, False])
        ])
print("5000 listened generados en data/listened.csv")
```

```python
# SONG → ARTIST
print("Generando song_artist")
with open("data/song_artist.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["song_id","artist_id"])

    for i in range(1, NUM_SONGS+1):
        writer.writerow([i, random.randint(1, NUM_ARTISTS)])
print(f"{NUM_SONGS} song_artist generados en data/song_artist.csv")
```

```python
# SONG → GENRE
print("Generando song_genre")
with open("data/song_genre.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["song_id","genre_id"])

    for i in range(1, NUM_SONGS+1):
        writer.writerow([i, random.randint(1, NUM_GENRES)])
print(f"{NUM_SONGS} song_genre generados en data/song_genre.csv")
```
