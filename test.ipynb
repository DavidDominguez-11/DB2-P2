```python
from faker import Faker
import csv
import random
import os

fake = Faker()

NUM_USERS = 1500
NUM_ARTISTS = 500
NUM_GENRES = 30
NUM_SONGS = 1500
NUM_PLAYLISTS = 1000
NUM_POSTS = 1000

genres_list = ["rock", "pop", "jazz", "hiphop", "electronic", "latin"]
devices = ["iPhone", "Android", "Web", "Desktop", "Tablet"]
reactions = ["love", "like", "wow", "sad", "angry"]

os.makedirs("data", exist_ok=True)

print("GENERANDO NODOS")
```

```python
print("Generando usuarios")
with open("data/users.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user_id","username","email","fecha_registro","premium","generos","es_artista"])
    for i in range(1, NUM_USERS+1):
        writer.writerow([
            i,
            fake.user_name(),
            fake.email(),
            fake.date_this_decade(),
            random.choice([True, False]),
            ",".join(random.sample(genres_list, k=random.randint(1,3))),
            random.random() < 0.1
        ])
print("Usuarios generados correctamente")
```

```python
print("Generando artistas")
with open("data/artists.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["artist_id","nombre","pais","genero_principal","activo","anios_activo"])
    for i in range(1, NUM_ARTISTS+1):
        writer.writerow([
            i,
            fake.name(),
            fake.country(),
            random.choice(genres_list),
            random.choice([True, False]),
            random.randint(1, 30)
        ])
print("Artistas generados correctamente")
```

```python
print("Generando generos")
with open("data/genres.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["genre_id","nombre","descripcion","popularidad","origen","activo"])
    for i in range(1, NUM_GENRES+1):
        writer.writerow([
            i,
            random.choice(genres_list),
            fake.sentence(),
            random.randint(1,100),
            fake.country(),
            True
        ])
print("Generos generados correctamente")
```

```python
print("Generando canciones")
with open("data/songs.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["song_id","titulo","duracion","fecha_lanzamiento","popularidad","idiomas"])
    for i in range(1, NUM_SONGS+1):
        writer.writerow([
            i,
            fake.word(),
            round(random.uniform(2,5),2),
            fake.date_this_decade(),
            random.randint(1,100),
            "english,spanish"
        ])
print("Canciones generadas correctamente")
```

```python
print("Generando playlists")
with open("data/playlists.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["playlist_id","nombre","descripcion","fecha_creacion","publica","numero_canciones","destacada"])
    for i in range(1, NUM_PLAYLISTS+1):
        writer.writerow([
            i,
            fake.word(),
            fake.sentence(),
            fake.date_this_decade(),
            random.choice([True, False]),
            random.randint(5,50),
            random.random() < 0.15
        ])
print("Playlists generadas correctamente")
```

```python
print("Generando posts")
with open("data/posts.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["post_id","caption","fecha","tipo","privacidad","hashtags","promocionado"])
    for i in range(1, NUM_POSTS+1):
        writer.writerow([
            i,
            fake.sentence(),
            fake.date_this_decade(),
            random.choice(["music","share","recommend"]),
            random.choice(["public","private"]),
            "#music,#fun",
            random.random() < 0.05
        ])
print("Posts generados correctamente")
```

```python
print(" GENERANDO RELACIONES")
```

```python
print("Generando FOLLOWS")
with open("data/follows.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user1","user2","fecha","notificaciones","cercania"])
    for i in range(1, NUM_USERS+1):
        u2 = random.choice([u for u in range(1, NUM_USERS+1) if u != i])
        writer.writerow([i, u2, fake.date_this_decade(), random.choice([True, False]), round(random.random(),2)])
        for _ in range(random.randint(0, 3)):
            u_rand = random.randint(1, NUM_USERS)
            if i != u_rand:
                writer.writerow([i, u_rand, fake.date_this_decade(), random.choice([True, False]), round(random.random(),2)])
print("FOLLOWS generado correctamente")
```

```python
print("Generando LISTENED")
with open("data/listened.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user_id","song_id","fecha","duracion_escuchada","completado"])
    for i in range(1, NUM_USERS+1):
        s_id = random.randint(1, NUM_SONGS)
        writer.writerow([i, s_id, fake.date_this_decade(), round(random.uniform(1,5),2), random.choice([True, False])])
        for _ in range(random.randint(1, 10)):
            writer.writerow([i, random.randint(1, NUM_SONGS), fake.date_this_decade(), round(random.uniform(1,5),2), random.choice([True, False])])
print("LISTENED generado correctamente")
```

```python
print("Generando BY")
with open("data/song_artist.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["song_id","artist_id","rol","fecha_colaboracion","principal"])
    for i in range(1, NUM_SONGS+1):
        writer.writerow([i, random.randint(1, NUM_ARTISTS), random.choice(["vocalist","producer","composer"]), fake.date_this_decade(), True])
        if random.random() < 0.2:
            writer.writerow([i, random.randint(1, NUM_ARTISTS), "feature", fake.date_this_decade(), False])
print("BY generado correctamente")
```

```python
print("Generando BELONGS_TO")
with open("data/song_genre.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["song_id","genre_id","relevancia","fecha_asignacion","principal"])
    for i in range(1, NUM_SONGS+1):
        writer.writerow([i, random.randint(1, NUM_GENRES), round(random.random(),2), fake.date_this_decade(), True])
print("BELONGS_TO generado correctamente")
```

```python
print("Generando CREATED (Posts)")
with open("data/created_posts.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user_id","post_id","fecha_creacion","dispositivo","ubicacion"])
    for i in range(1, NUM_POSTS+1):
        writer.writerow([random.randint(1, NUM_USERS), i, fake.date_this_decade(), random.choice(devices), fake.city()])
print("CREATED generado correctamente")
```

```python
print("Generando CONTAINS")
with open("data/post_songs.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["post_id","song_id","orden","destacado","timestamp"])
    for i in range(1, NUM_POSTS+1):
        writer.writerow([i, random.randint(1, NUM_SONGS), 1, True, round(random.uniform(0,30),2)])
print("CONTAINS generado correctamente")
```

```python
print("Generando LIKED")
with open("data/liked.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user_id","post_id","fecha","reaccion","intensidad"])
    for _ in range(3000):
        writer.writerow([random.randint(1, NUM_USERS), random.randint(1, NUM_POSTS), fake.date_this_decade(), random.choice(reactions), random.randint(1,10)])
print("LIKED generado correctamente")
```

```python
print("Generando COMMENTED")
with open("data/commented.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user_id","post_id","fecha","contenido","likes"])
    for _ in range(2000):
        writer.writerow([random.randint(1, NUM_USERS), random.randint(1, NUM_POSTS), fake.date_this_decade(), fake.sentence(), random.randint(0,100)])
print("COMMENTED generado correctamente")
```

```python
print("Generando CREATED_PLAYLIST")
with open("data/created_playlists.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user_id","playlist_id","fecha","colaborativa","numero_ediciones"])
    for i in range(1, NUM_PLAYLISTS+1):
        writer.writerow([random.randint(1, NUM_USERS), i, fake.date_this_decade(), random.choice([True, False]), random.randint(1,20)])
print("CREATED_PLAYLIST generado correctamente")
```

```python
print("Generando INCLUDES")
with open("data/playlist_songs.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["playlist_id","song_id","fecha_agregado","orden","favorito"])
    for i in range(1, NUM_PLAYLISTS+1):
        for order in range(1, random.randint(5, 15)):
            writer.writerow([i, random.randint(1, NUM_SONGS), fake.date_this_decade(), order, random.choice([True, False])])
print("INCLUDES generado correctamente")
```

```python
print("Generando FOLLOWS_ARTIST")
with open("data/follows_artist.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user_id","artist_id","fecha","notificaciones","interacciones"])
    for _ in range(2500):
        writer.writerow([random.randint(1, NUM_USERS), random.randint(1, NUM_ARTISTS), fake.date_this_decade(), random.choice([True, False]), random.randint(0,50)])
print("FOLLOWS_ARTIST generado correctamente")
```

```python
print("Generando SAVED")
with open("data/saved_playlists.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user_id","playlist_id","fecha","veces_reproducida","favorita"])
    for _ in range(2000):
        writer.writerow([random.randint(1, NUM_USERS), random.randint(1, NUM_PLAYLISTS), fake.date_this_decade(), random.randint(0,100), random.choice([True, False])])
print("SAVED generado correctamente")
```
